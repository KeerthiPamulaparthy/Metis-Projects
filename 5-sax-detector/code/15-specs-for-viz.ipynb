# Spectrograms for Viz

```python
import numpy as np
import audiomod
import librosa
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
def spectro_generator(chunk_id, fig_dims, margin=True, save=False):
    """
    Saves or shows a spectrogram.
    ---
    chunk_id: six-digit chunk ID (str)
    fig_dims: tuple of figure size in inches (tuple of ints)
    save: if True, will save in 'specs' directory; if False, will show
    """
    y, sr = audiomod.audio_loader(chunk_id)
    spec = audiomod.make_spectro(y, sr)
    audiomod.spectro_viz(spec, save=save, chunk_id=chunk_id, margin=margin, fig_dims=fig_dims)
```

```python
spectro_generator('002767', fig_dims=(16, 9), margin=False, save=True)
```

```python
spectro_generator('cowbell', fig_dims=(8, 8), margin=False, save=True)
```
